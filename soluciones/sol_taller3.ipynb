```python
import pandas as pd
```

```python
#ruta_archivo = '/content/consumo_masivo_col.xlsx' #Para GoogleColab
ruta_archivo = '../clase_3/data/consumo_masivo_col.xlsx' #Para local
productos_consumo_masivo = pd.read_excel(ruta_archivo)
```

### Trabajando con fechas al importar un archivo de excel

```python
productos_consumo_masivo['date'] = pd.to_datetime(productos_consumo_masivo['date'],format='%Y%m%d')
```

### Trabajando con valores numéricos al importar un archivo de excel

```python
productos_consumo_masivo['prod_unit_price'] = pd.to_numeric(productos_consumo_masivo['prod_unit_price'].astype(str).str.replace(',', '.'),errors='coerce')
```

A partir del dataframe anterior, responda las siguientes preguntas:

1. Antes de iniciar, trate de familiarizarse un poco con la base de datos. ¿Cuántos valores de `subcategory` existen? ¿A qué corresponden? ¿Cuántos valores de `prod_units` existen? ¿A qué corresponden?

```python
subcategorias = productos_consumo_masivo['subcategory'].unique()
print(f'Hay un total de {len(subcategorias)} subcategorías que corresponden a las categorías de los productos: {list(subcategorias)}.')
```

```python
prod_units = productos_consumo_masivo['prod_units'].unique()
print(f'Hay un total de {len(prod_units)} valores de prod_units que corresponden a la unidad del precio: {list(prod_units)}.')
```

2. ¿Cuál es el rango de precios que se encuentra en esta base de datos?

```python
min_price = productos_consumo_masivo['prod_unit_price'].min()
max_price = productos_consumo_masivo['prod_unit_price'].max()
print(f'Los precios oscilan entre los ${min_price:,.0f} y ${max_price:,.0f} COP.')
```

3. ¿Cuál es el precio promedio de los artículos de cuidado personal?

```python
prod_cuidado_personal = productos_consumo_masivo.loc[productos_consumo_masivo['subcategory']=='Cuidado Personal'].drop_duplicates(subset='prod_id')
precio_promedio_cuidado_personal = prod_cuidado_personal['prod_unit_price'].mean()
print(f'El precio promedio de los artículos de cuidado personal es de ${round(precio_promedio_cuidado_personal,-2):,} COP.')
prod_cuidado_personal.head()
```

4. ¿Cuáles son  las cinco `Galletas dulces` más costosas del mercado? ¿A qué marcas corresponden?

```python
galletas_dulces = productos_consumo_masivo.loc[productos_consumo_masivo['tags']=='Galletas dulces'].drop_duplicates(subset='prod_id')
galletas_mas_costosas = galletas_dulces.sort_values(by='prod_unit_price',ascending=False).head(5)
print(f'Las cinco galletas más costosas son: {', '.join(galletas_mas_costosas.prod_name.values)} que corresponden a las marcas {', '.join(galletas_mas_costosas.prod_brand.unique())}')
galletas_mas_costosas
```

6. Construya el dataframe con las desviaciones estándar de los precios promedios y desviaciones estándar de todas las **botellas** de licor en el mercado? El resultado debe ser algo similar a la siguiente tabla:

| **tags**    | **promedio** | **std**      |
|-------------|--------------|--------------|
| Aguardiente | 34904.14966  | 19728.620641 |
| Aperitivos  | 38928.155738 | 32223.129149 |
| ...         | ...          | ...          |
| Whisky      | 97422.382114 | 62723.513287 |

```python
botellas_licor = productos_consumo_masivo.loc[(productos_consumo_masivo.subcategory=='Vinos y Licores')&(productos_consumo_masivo.prod_units=='Botella')].drop_duplicates(subset='prod_id')
botellas_licor.head()
```

```python
tipos_licor = botellas_licor.tags.unique()
promedios = [float(botellas_licor.loc[botellas_licor.tags==tipo,'prod_unit_price'].mean()) for tipo in tipos_licor]
std = [float(botellas_licor.loc[botellas_licor.tags==tipo,'prod_unit_price'].std()) for tipo in tipos_licor]

pd.DataFrame({'tags':tipos_licor,'promedios':promedios,'std':std})
```

7. ¿Cuántos productos tienen como fecha Febrero de 2020? ¿Cuáles son las dos marcas con más productos en esta fecha?

```python
prod_febrero2020 = productos_consumo_masivo.loc[(productos_consumo_masivo.date>=pd.to_datetime('2020-02-01')) & (productos_consumo_masivo.date<pd.to_datetime('2020-03-01'))].drop_duplicates(subset='prod_id')
print(f'Hay un total de {prod_febrero2020.shape[0]:,} productos registrados en Febrero de 2020.')
prod_febrero2020.head()
```
